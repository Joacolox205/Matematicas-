```python
#pregunta 1

import numpy as np
from scipy.optimize import fsolve

def S(x):
    return -1.6*x**2+1.3*x+141.3
def R(x):
    return 1.4*x**2-5.6*x+12.6

def inter(x):
    return S(x)-R(x)

xo = np.linspace(-7,20,2)
solucion = fsolve(inter, xo)
print(f'{solucion}')
```

```python
#Pregunta 3
import numpy as np
from scipy.optimize import fsolve

def D(t):
  return 50*np.exp(0.3*t)-551

xo = np.linspace(0,20,1)
solucion = fsolve(D, xo)
print(f'{solucion[0]:.0f}')
```

```python
#pregunta 15
import numpy as np
from scipy.optimize import fsolve

def I(x):
    return 50*x+0.5*x**2
def C(x):
    return 10*x+450

def inter(x):
    return I(x)-C(x)

xo = np.linspace(0,20,1)
solucion = fsolve(inter, xo)
print(f'{solucion[0]:.0f}')
```
