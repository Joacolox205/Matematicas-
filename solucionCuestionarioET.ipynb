```python
#La forma algebraica de la función que relaciona ambas variables es:
import numpy as np
x = np.array([5,10,15,20])
y = np.array([200,300,400,500])

pendiente, intercepto = np.polyfit(x, y, 1)

print(f'la pendiente es: {pendiente} y el coeficiente de posicion es: {intercepto}')
```

```python
#¿Cuántos años deben pasar que la base de datos tenga un tamaño de 20 TB?
from scipy.optimize import fsolve

def D(t):
    return 2*1.2**t-20
xo = np.linspace(0, 20, 1)
solucion = fsolve(D,xo)
print(f'{solucion}')
```

```python
#Un algoritmo demora en procesar un solo registro en 2,1 minutos; dos registros en 2,2 y tres registros en 2,3 minutos. Si se sigue el mismo patrón de crecimiento, ¿cuántos registros procesa el algoritmo en 5,6 minutos?
#opcion 1
final = 5.6
termino = 0
dif = final - termino
n = 0
a_1 =2.1
d=0.1
while dif > 0:
  n += 1
  termino = a_1+(n-1)*d
  dif = final - termino
print(f'El término 5.6 se ubica en la posición {n} de la sucesión.')
```

```python
#En una tienda recién inaugurada, la primera semana ha vendido 300 unidades de un producto determinado y se espera que las ventas aumenten en 40 unidades cada semana. ¿Cuántas unidades del producto se proyecta vender en total al cabo de 2 meses? Considere que 1 mes equivale a 4 semanas.

a_1=300
d=40
sucesion = []
for i in range(8):
    n = i + 1
    sucesion.append(a_1+(n-1)*d)
suma = sum(sucesion)
print(f'{suma}')
```

```python
#En una sucesión geométrica, el cuarto término es 24 y el sexto término es 96. Determine el valor del noveno término.
# El noveno término es 
a_4=24
a_6=96

r=(a_6/a_4)**(1/2)
print(f'r={r}')
a_1=a_4/r**3
print(f'a_1={a_1}')
n=9
a_9=a_1*r**(n-1)
print(f'a_9={a_9}')
```

```python
#¿En cuánto está evaluado, para el miércoles de la semana pasada, el total de acciones que posee en esta empresa el inversionista 8?

A = np.fromfunction(lambda i, j: 20*(i+1)*(j+1), (15,10))
E = np.fromfunction(lambda i, j: 4*(i+1)+(j+1), (10,5))

C=np.dot(A,E)
print(f'{C}')
```
