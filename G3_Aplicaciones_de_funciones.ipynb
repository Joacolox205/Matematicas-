# Problema 1

Sean las funciones:

\begin{split}
r(x) & = -1,5x^{2} + 11,5x - 15\\
g(x) & = 1,58x^{4} - 19,17x^{3} + 80,92x^{2} - 139,33x + 85
\end{split}

<br>

Considerando que $\hspace{0.2cm}$ $0 \leq x \leq 6$, $\hspace{0.3cm}$  $-10 \leq y \leq 10$ $\hspace{0.2cm}$ realice lo siguiente:

1. Grafique ambas funciones considerando el dominio indicado. Sugerencia: utilice *plt.xlim()* y *plt.ylim()* para definir los límites de los ejes en el gráfico.
2. ¿Cuáles son los puntos de intersección entre las funciones? Utilice *fsolve* para encontrar las soluciones.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# --- Problema 1 ---
print("### Problema 1 ###")

# 1. Definición y Gráfico de Funciones
def r(x):
    return -1.5*x**2 + 11.5*x - 15

def g(x):
    return 1.58*x**4 - 19.17*x**3 + 80.92*x**2 - 139.33*x + 85

#
x = np.arange(0, 6.01, 0.01) # Dominio 0 <= x <= 6
plt.plot(x, r(x), label = 'r(x)')
plt.plot(x, g(x), label = 'g(x)')
plt.title('Gráfico Problema 1')
plt.ylabel('Eje Y')
plt.xlabel('Eje X')
plt.legend()
plt.grid(True)
# Sugerencia: definir límites de ejes
plt.ylim(-10, 10)
plt.xlim(0, 6)
plt.show()

# 2. Puntos de intersección
#
def inter_1(x):
    return r(x) - g(x) # 'función igualada a cero'

#
# Buscamos soluciones en el dominio [0, 6]
xo_1 = np.linspace(0, 6, 3) # Buscamos 3 puntos iniciales
solucion_1 = fsolve(inter_1, xo_1)

print("\n2. Puntos de intersección (valores de x):")
# Filtramos soluciones que están dentro del dominio y son únicas
soluciones_unicas_1 = np.unique(np.round(solucion_1[solucion_1 >= 0], 4))
print(soluciones_unicas_1)
```

# Problema 2

Sean las funciones:

\begin{split}
h(x) & = 2(x - 2)^{2}\\
i(x) & = x + 1
\end{split}

<br>

1. Grafique ambas funciones en un mismo gráfico utilizando *Matplotlib*.
2. ¿En qué punto(s) interseca la función $h(x)$ al eje $x$? ¿y la función $i(x)$?
3. ¿Las gráficas de las funciones se intersecan en algún punto?
4. ¿Para qué valores de $x$ ocurre que $h(x) \leq i(x)$?
5. Determine el punto en que $h(x) = 50$.


```python
# --- Problema 2 ---
print("\n### Problema 2 ###")

def h(x):
    return 2 * (x - 2)**2

def i(x):
    return x + 1

# 1. Gráfico de ambas funciones
x = np.arange(-2, 6, 0.1) # Rango ampliado para ver las intersecciones
plt.plot(x, h(x), label = 'h(x)')
plt.plot(x, i(x), label = 'i(x)')
plt.title('Gráfico Problema 2')
plt.ylabel('Eje Y')
plt.xlabel('Eje X')
plt.legend()
plt.grid(True)
plt.show()

# 2. Intersecciones con el eje X (f(x) = 0)
#
sol_h = fsolve(h, 0) # h(x) = 0
sol_i = fsolve(i, 0) # i(x) = 0
print("\n2. Intersecciones con eje X:")
print(f"h(x) interseca el eje x en: {sol_h[0]}")
print(f"i(x) interseca el eje x en: {sol_i[0]}")

# 3. Intersección entre funciones
def inter_2(x):
    return h(x) - i(x) # h(x) = i(x)

solucion_2 = fsolve(inter_2, [0, 3]) # Damos dos puntos iniciales (0 y 3)
print("\n3. Puntos de intersección (valores de x):")
print(solucion_2)

# 4. Intervalo h(x) <= i(x)
print("\n4. Intervalo donde h(x) <= i(x):")
print(f"Viendo el gráfico, h(x) está por debajo de i(x) entre los puntos de intersección.")
print(f"El intervalo es: [{solucion_2[0]:.2f}, {solucion_2[1]:.2f}]")

# 5. Determinar punto donde h(x) = 50
def h_50(x):
    return h(x) - 50 # 'función igualada a cero'

solucion_50 = fsolve(h_50, [-10, 10]) # Buscamos 2 soluciones (una negativa y una positiva)
print("\n5. Puntos donde h(x) = 50 (valores de x):")
print(solucion_50)
```

# Problema 3

Sean las funciones:

\begin{split}
w(x) & = 0,6x^{2} + 1,5\\
t(x) & = 0,8x^{3} - 10x + 10
\end{split}

<br>

Considerando que $\hspace{0.2cm}$ $-3 \leq x \leq 4$, $\hspace{0.2cm}$ $-4 \leq y \leq 24$ $\hspace{0.2cm}$ realice lo siguiente:

1. Grafique ambas funciones en un mismo gráfico, indicando unidad de medida. Utilice *Matplotlib*.
2. Determine el intervalo en que $w(x) \geq t(x)$.
3. Encuentre los puntos en que $t(x) = 6$.


```python
# --- Problema 3 ---
print("\n### Problema 3 ###")

def w(x):
    return 0.6*x**2 + 1.5

def t(x):
    return 0.8*x**3 - 10*x + 10

# 1. Gráfico de ambas funciones
x = np.arange(-3, 4.01, 0.01) # Dominio -3 <= x <= 4
plt.plot(x, w(x), label = 'w(x)')
plt.plot(x, t(x), label = 't(x)')
plt.title('Gráfico Problema 3')
plt.ylabel('Unidad Y')
plt.xlabel('Unidad X')
plt.legend()
plt.grid(True)
# Límites de los ejes dados
plt.ylim(-4, 24)
plt.xlim(-3, 4)
plt.show()

# 2. Intervalo w(x) >= t(x)
def inter_3(x):
    return w(x) - t(x)

#
xo_3 = np.array([-3, 0, 3]) # Puntos iniciales basados en el gráfico
solucion_3 = fsolve(inter_3, xo_3)
soluciones_unicas_3 = np.unique(np.round(solucion_3, 4))
print("\n2. Puntos de intersección (valores de x):")
print(soluciones_unicas_3)
print("Viendo el gráfico, w(x) es mayor o igual a t(x) en los intervalos:")
print(f"[{soluciones_unicas_3[0]:.2f}, {soluciones_unicas_3[1]:.2f}] y [{soluciones_unicas_3[2]:.2f}, 4.00]")

# 3. Puntos en que t(x) = 6
def t_6(x):
    return t(x) - 6 # 'función igualada a cero'

xo_t6 = np.array([-3, 1, 3]) # Puntos iniciales basados en el gráfico
solucion_t6 = fsolve(t_6, xo_t6)
soluciones_unicas_t6 = np.unique(np.round(solucion_t6, 4))
print("\n3. Puntos donde t(x) = 6 (valores de x):")
print(soluciones_unicas_t6)
```

Te contrataron como asesor en una campaña de marketing digital y necesitas equilibrar el ingreso por ventas con los costos de inversión.

Las funciones de ingresos por ventas $I(x)$ y costos de inversión $C(x)$ (ambas en miles de pesos) son:

<br>

$$I(x) = - 20x + 1.000$$

$$C(x) = 200x + 500$$

<br>

Donde $x$ corresponde al tiempo transcurrido desde que se inició la campaña (en semanas).

<br>

1. Analice el comportamiento del ingreso por ventas y del costo de inversión a medida que aumenta el tiempo desde que se inició la campaña. Considere la pendiente y el intercepto de cada función para realizar el análisis.
2. Grafique ambas funciones en un mismo gráfico, indicando nombre
de los ejes. Utilice *Matplotlib*.
3. Determine, utilizando *Python*, el punto de equilibrio entre ambas funciones e interprete el resultado.
4. Utilizando el gráfico y lo realizado en el punto $3$, ¿qué puedes concluir? ¿Aconsejarías mantener la campaña transcurridas $5$ semanas desde el inicio de la campaña?

```python
# --- Problema 5 ---
print("\n### Problema 5 ###")

def I(x):
    return -20 * x + 1000

def C(x):
    return 200 * x + 500

# 2. Gráfico
x = np.arange(0, 10, 0.1) # Graficamos 10 semanas
plt.plot(x, I(x), label = 'Ingresos I(x)')
plt.plot(x, C(x), label = 'Costos C(x)')
plt.title('Ingresos vs Costos de Campaña')
plt.ylabel('Monto (miles de pesos)')
plt.xlabel('Tiempo (semanas)')
plt.legend()
plt.grid(True)
plt.show()

# 3. Punto de equilibrio (I(x) = C(x))
def equilibrio(x):
    return I(x) - C(x)

#
solucion_eq = fsolve(equilibrio, 0)
print(f"\n3. Punto de equilibrio:")
print(f"Los costos igualan a los ingresos a las {solucion_eq[0]:.2f} semanas.")
print("Interpretación: En este punto, la campaña deja de generar ganancias y empieza a generar pérdidas.")

# 4. Conclusión
print("\n4. Conclusión:")
print(f"El punto de equilibrio es aprox. 2.27 semanas. Después de eso, los costos (línea naranja) son siempre mayores que los ingresos (línea azul).")
print("No aconsejaría mantener la campaña 5 semanas, ya que estaría generando grandes pérdidas.")
```

# Problema 8

Constanza es agente inmobiliario, el sueldo en su actual trabajo depende de la cantidad de casas vendidas, le pagan $\$600.000$ de sueldo base, más una comisión del $5\%$ por cada casa vendida. Lo anterior puede modelarse mediante la función $S_{actual}(x)$ (en miles de pesos):

<br>

$$S_{actual}(x) = 600 \cdot 1,05^{x}$$

<br>

donde $x$ corresponde a la cantidad de casas vendidas durante el mes. Debido al aumento en el costo de la vida, necesita aumentar sus ingresos. Luego de postular a varios trabajos, recibió una oferta laboral donde le ofrecen un sueldo base de $\$500.000$ más una comisión de venta del $7\%$ por cada casa vendida.

<br>

Considerando el contexto anterior encuentre:

1. La forma algebraica que modela el sueldo mensual de la oferta recibida.
2. Grafique ambas funciones usando *Matplotlib*. Considera que vende un máximo de $30$ casas mensuales.
3. Si tuvieras que aconsejar a Constanza, ¿le recomendarías que tome la oferta o que se mantenga en su actual trabajo? Justifica matemáticamente.

```python
# --- Problema 8 ---
print("\n### Problema 8 ###")

def S_actual(x):
    return 600 * (1.05**x)

def S_nueva(x):
    return 500 * (1.07**x)

# 2. Gráfico
x = np.arange(0, 31, 1) # Máximo 30 casas
plt.plot(x, S_actual(x), label = 'Sueldo Actual (5%)')
plt.plot(x, S_nueva(x), label = 'Sueldo Nuevo (7%)')
plt.title('Comparación de Sueldos')
plt.ylabel('Sueldo (miles de pesos)')
plt.xlabel('Casas vendidas')
plt.legend()
plt.grid(True)
plt.show()

# 3. Aconsejar a Constanza (encontrar intersección)
def inter_8(x):
    return S_actual(x) - S_nueva(x)

#
solucion_8 = fsolve(inter_8, 10)
print(f"\n3. Consejo:")
print(f"El punto de cruce ocurre al vender {solucion_8[0]:.2f} casas.")
print("Recomendación: Si Constanza vende 9 casas o menos, debe quedarse en su trabajo actual. Si vende 10 casas o más, debe tomar la nueva oferta.")
```

En un laboratorio de biología marina, los científicos están estudiando el crecimiento de dos especies de algas en función de la luz solar recibida. Las funciones $f(x)$ y $g(x)$, que describen la tasa de crecimiento de las algas en milímetros por día (mm/día) son:

<br>

$$f(x) = \sqrt{500 x + 2.000}$$

<br>

$$g(x) = \sqrt{800 x + 1.000}$$

<br>

donde $x$ corresponde a la intensidad de la luz (en lux).

<br>

1. Determina el punto en el que ambas especies de algas tienen la misma tasa de crecimiento. Además, calcula la tasa de crecimiento en ese punto. Visualiza apropiadamente.
2. Considerando lo obtenido en el ítem anterior, analice y compare el crecimiento de las dos especies en función de la intensidad de la luz.

```python
# --- Problema 9 ---
print("\n### Problema 9 ###")

def f(x): # Especie 1
    return np.sqrt(500 * x + 2000)

def g(x): # Especie 2
    return np.sqrt(800 * x + 1000)

# 1. Punto de misma tasa de crecimiento
def inter_9(x):
    return f(x) - g(x)

#
solucion_9 = fsolve(inter_9, 0)
print(f"1. Ambas especies tienen la misma tasa de crecimiento a {solucion_9[0]:.2f} lux.")

# Calcular la tasa en ese punto
tasa_crecimiento = f(solucion_9[0])
print(f"   La tasa de crecimiento en ese punto es de {tasa_crecimiento:.2f} mm/día.")

# Gráfico para visualizar
x = np.arange(0, 10, 0.1)
plt.plot(x, f(x), label = 'Especie 1 (f(x))')
plt.plot(x, g(x), label = 'Especie 2 (g(x))')
plt.title('Crecimiento de Algas vs. Intensidad de Luz')
plt.ylabel('Tasa de Crecimiento (mm/día)')
plt.xlabel('Intensidad de Luz (lux)')
plt.legend()
plt.grid(True)
plt.show()

# 2. Análisis
print("\n2. Análisis:")
print(f"Viendo el gráfico, si la intensidad de luz es menor a {solucion_9[0]:.2f} lux, la Especie 1 (f(x)) crece más rápido.")
print(f"Si la intensidad es mayor a {solucion_9[0]:.2f} lux, la Especie 2 (g(x)) crece más rápido.")
```

Sea la función $f$, definida por:

```python
import numpy as np
from numpy.linalg import inv # Importamos inv del formulario

# --- Problema 11 ---
print("### Problema 11: Monedas ###")

# 1. Definimos la matriz de coeficientes A y la matriz de resultados B
A = np.array([
    [ 1,  1],
    [10, 50]
])

B = np.array([
    [200],
    [8000]
])

# 2. Calculamos la inversa de A
A_inv = inv(A) #

# 3. Calculamos la solución X = A_inv * B
Solucion = np.dot(A_inv, B) #

print("3. Cantidad de Monedas:")
# Usamos .round() para asegurarnos de obtener enteros
print(f"Tiene {int(np.round(Solucion[0,0]))} monedas de $10 (x)")
print(f"Tiene {int(np.round(Solucion[1,0]))} monedas de $50 (y)")
```
