# Problema 1

Las matrices $A$ y $B$ representan las ventas en dos años, en millones de dólares, de los tres productos de una empresa en las cuatro regiones en que está disponible. Es decir, $a_{ij}$ y $b_{ij}$ representan el monto de las ventas del producto $i$ en la región $j$ el primer y segundo año, respectivamente:

<br>

\begin{split}
A = \begin{pmatrix}
 2,6 &  4,8 &  1,8 & 0,9 \\
 3,2 &  4,4 &  2,5 & 2,8 \\
 2,4 &  3,6 &  3,8 & 2,5 \\
\end{pmatrix} \text{ } B= \begin{pmatrix}
 3,6 &  2,5 &  3,0 & 2,5 \\
 4,5 &  5,0 &  3,5 & 3,8 \\
 2,9 &  3,0 &  4,6 & 4,0 \\
\end{pmatrix}
\end{split}

<br>

1. Indique el orden de cada matriz.
2. Calcule la matriz $S = A + B$.
3. Calcule la matriz $D = B - A$.
4. Indique el valor e interprete los elementos $s_{24}$ y $d_{24}$.
5. Para el tercer año de funcionamiento, los tres productos ahora están disponibles en seis regiones. Durante este año, las ventas de esta empresa estarán registradas en una matriz $E$. ¿Es posible determinar la matriz $T = A + B + E$? Justifique su respuesta.

```python
import numpy as np

# Definimos las matrices A y B
A = np.array([
    [2.6, 4.8, 1.8, 0.9],
    [3.2, 4.4, 2.5, 2.8],
    [2.4, 3.6, 3.8, 2.5]
])

B = np.array([
    [3.6, 2.5, 3.0, 2.5],
    [4.5, 5.0, 3.5, 3.8],
    [2.9, 3.0, 4.6, 4.0]
])

# 2. Calcular S = A + B
S = A + B
print("--- Matriz S (Ventas Totales) ---")
print(S)

# 3. Calcular D = B - A
D = B - A
print("\n--- Matriz D (Diferencia de Ventas) ---")
print(D)

# 4. Interpretar s_24 y d_24
# En Python, los índices empiezan en 0, así que [1, 3] es la fila 2, columna 4.
s_24 = S[1, 3]
d_24 = D[1, 3]

print(f"\nValor de s_24: {s_24:.1f}")
print(f"Valor de d_24: {d_24:.1f}")
```

# Problema 3

La empresa Beta tiene cinco plantas en Chile, y en cada una de ellas fabrica siete productos. El número de unidades del producto $i$ fabricadas en la planta $j$ en una semana está representado en la matriz $C$ en el elemento $c_{ij}$.

<br>

\begin{split}
C = \begin{pmatrix}
 120 &  200 &  155 &  125  &  230 \\
 155 &  130 &  110 &  160  &  180 \\
 140 &  205 &  150 &  125  &  155 \\
 130 &  120 &  230 &  225  &  125 \\
 135 &  70  &  190 &  175  &  95  \\
 110 &  265 &  180 &  230  &  180 \\
 185 &  85  &  120 &  135  &  150 \\
\end{pmatrix}
\end{split}

<br>

La empresa se expande, creando cinco plantas en Ecuador que producen los mismos siete productos. La producción semanal de esta filial estará representada por una matriz $E$.

1. Si se sabe que los volúmenes de producción semanal de la filial en Ecuador son un $20\%$ menor que en Chile, calcule la matriz $E$ utilizando *Python*.
2. Si representamos la producción semanal total de la empresa Beta, considerando ambas filiales, por la matriz $T$, escriba la expresión que calcula cada elemento de la matriz $T$ usando notación matricial.
3. Calcule la matriz $T$ utilizando *Python*.
4. Interprete los elementos $c_{34}$ y $t_{72}$ en el contexto del problema.

```python
import numpy as np

# Matriz C (Chile)
C = np.array([
    [120, 200, 155, 125, 230],
    [155, 130, 110, 160, 180],
    [140, 205, 150, 125, 155],
    [130, 120, 230, 225, 125],
    [135,  70, 190, 175,  95],
    [110, 265, 180, 230, 180],
    [185,  85, 120, 135, 150]
])

# 1. Calcular la matriz E (Producción Ecuador)
E = 0.8 * C
print("--- 1. Matriz E (Producción Ecuador) ---")
# Usamos .astype(int) para mostrar números enteros, ya que no se pueden fabricar unidades decimales.
print(E.astype(int))

# 2. Expresión matricial para T
print("\n--- 2. Expresión Matricial ---")
print("La expresión que calcula T es: T = C + E")

# 3. Calcular la matriz T (Producción Total)
T = C + E
print("\n--- 3. Matriz T (Producción Total) ---")
print(T)

# 4. Interpretación de elementos
# Recordamos que Python usa índices desde 0.
# c_34 es fila 3, columna 4 -> índice [2, 3]
# t_72 es fila 7, columna 2 -> índice [6, 1]

c_34 = C[2, 3]
t_72 = T[6, 1]

print("\n--- 4. Interpretación de Elementos ---")
print(f"El elemento c_34 ({c_34}) representa la producción de 125 unidades del producto 3 en la planta 4 de la filial en Chile.")
print(f"El elemento t_72 ({t_72}) representa la producción total de 153 unidades del producto 7 en la planta 2, sumando las filiales de Chile y Ecuador.")
```

# Problema 4

Una empresa que fabrica zapatos tiene dos plantas, una en Valparaíso y la otra en Temuco. La empresa produce zapatos de color negro, blanco y café, tanto para niños, damas como caballeros. La capacidad actual de producción mensual (en miles de pares) para Valparaíso y Temuco, están dadas por las matrices $V$ y $T$, respectivamente. Para ambas matrices, según el orden en que se mencionan, las filas corresponden al color del calzado y las columnas al tipo de persona que utilizará el calzado.

<br>

\begin{split}
V = \begin{pmatrix}
 30 &  34 &  20  \\
 45 &  20 &  16  \\
 14 &  26 &  25  \\
\end{pmatrix} \text{ } T = \begin{pmatrix}
 35 &  30 &  26  \\
 52 &  25 &  18  \\
 23 &  24 &  32  \\
\end{pmatrix}
\end{split}

<br>

1. Determine la producción mensual total de ambas plantas, representándola en una matriz $A$. Utilice *Python* para resolver.
2. Interprete el elemento $a_{21}$ en el contexto del problema.
3. Si para el próximo año se estima que la producción en Valparaíso aumenta en un $30\%$, mientras que la de Temuco se reduce en un $10\%$, represente la nueva producción total en la matriz $B$.
4. Interprete el elemento $b_{13}$ en el contexto del problema.

```python
import numpy as np

# Matriz V (Valparaíso) y T (Temuco)
V = np.array([
    [30, 34, 20],
    [45, 20, 16],
    [14, 26, 25]
])

T = np.array([
    [35, 30, 26],
    [52, 25, 18],
    [23, 24, 32]
])

# 1. Producción mensual total A
A = V + T
print("--- 1. Matriz A (Producción Total Actual) ---")
print(A)

# 2. Interprete a_21
# Fila 2, Columna 1 -> índice [1, 0]
a_21 = A[1, 0]
print("\n--- 2. Interpretación a_21 ---")
print(f"a_21 ({a_21}) representa la producción total mensual de 97 mil pares de zapatos de color Blanco para Niños.")

# 3. Nueva producción B (Valparaíso +30%, Temuco -10%)
V_nueva = V * 1.30
T_nueva = T * 0.90
B = V_nueva + T_nueva
print("\n--- 3. Matriz B (Nueva Producción Total) ---")
print(B)

# 4. Interprete b_13
# Fila 1, Columna 3 -> índice [0, 2]
b_13 = B[0, 2]
print("\n--- 4. Interpretación b_13 ---")
print(f"b_13 ({b_13:.1f}) representa la producción total mensual esperada de 49.4 mil pares de zapatos de color Negro para Caballeros.")
```

# Problema 5

Genere una matriz $M$ de orden $8 \times 9$ donde cada término de la matriz está definido por:

$$m_{ij} = 8i + 12j$$

y calcule su transpuesta.

```python
import numpy as np

# Usamos np.fromfunction. (8, 9) es el orden de la matriz.
# i tomará valores de 0 a 7.
# j tomará valores de 0 a 8.
M = np.fromfunction(lambda i, j: 8*i + 12*j, (8, 9), dtype=int)

print("--- Matriz M (m_ij = 8i + 12j) ---")
print(M)

# Calcular la transpuesta usando .T
M_transpuesta = M.T

print("\n--- Matriz Transpuesta (M.T) ---")
print(M_transpuesta)
```

# Problema 8

Una empresa tecnológica vende software y hardware en tres productos: $P_{1}$, $P_{2}$ y $P_{3}$, en cuatro zonas de la capital: norte, sur, este y oeste. Las ventas del mes, en millones de pesos, están registradas en las matrices $S$ y $H$, para software y hardware respectivamente.

<br>

\begin{split}
S = \begin{pmatrix}
 260 &  480 &  180  &  90  \\
 320 &  440 &  250  &  280 \\
 240 &  360 &  380  &  250  \\
\end{pmatrix} \text{ } H = \begin{pmatrix}
 360 &  250 &  300  &  250  \\
 450 &  500 &  350  &  380 \\
 290 &  300 &  460  &  400  \\
\end{pmatrix}
\end{split}

<br>

1. Calcule en la matriz $T$, del total de ventas de la empresa tecnológica.
2. Un trabajador indica que este mes, la sucursal de la zona sur recaudó más por la venta hardware que software, ¿es correcta su afirmación?
3. Debido al black friday se espera que el siguiente mes, las ventas de software aumenten un $7\%$ y las de hardware aumenten un $3\%$ respecto al mes actual. Encuentra en total de ventas esperadas para el siguiente mes y almacénalo en la matriz $M$. Utiliza *Python* para realizar los cálculos.
4. Interpreta los elementos $t_{24}$ y $m_{31}$ en el contexto del problema.
5. Calcula la matriz $M^{T}$ e interpreta el elemento $m_{41}^{T}$ en el contexto del problema.


```python
import numpy as np

# Matriz S (Software) y H (Hardware)
S = np.array([
    [260, 480, 180,  90],
    [320, 440, 250, 280],
    [240, 360, 380, 250]
])

H = np.array([
    [360, 250, 300, 250],
    [450, 500, 350, 380],
    [290, 300, 460, 400]
])

# 1. Calcule T (Ventas Totales)
T = S + H
print("--- 1. Matriz T (Ventas Totales) ---")
print(T)

# 2. Afirmación del trabajador (Zona Sur = Columna índice 1)
# Usamos np.sum(matriz[:, indice_columna]) para sumar la columna
ventas_software_sur = np.sum(S[:, 1])
ventas_hardware_sur = np.sum(H[:, 1])

print("\n--- 2. Afirmación del Trabajador (Zona Sur) ---")
print(f"Ventas Software (Sur): {ventas_software_sur} millones")
print(f"Ventas Hardware (Sur): {ventas_hardware_sur} millones")
if ventas_hardware_sur > ventas_software_sur:
    print("La afirmación es CORRECTA.")
else:
    print("La afirmación es INCORRECTA. La zona sur recaudó más por Software (1280) que por Hardware (1050).")

# 3. Calcule M (Ventas mes siguiente: S +7%, H +3%)
M = (S * 1.07) + (H * 1.03)
print("\n--- 3. Matriz M (Ventas Siguiente Mes) ---")
print(np.round(M, 2))

# 4. Interprete t_24 y m_31
# t_24 -> Fila 2, Columna 4 de T
t_24 = T[1, 3]
# m_31 -> Fila 3, Columna 1 de M
m_31 = M[2, 0]
print("\n--- 4. Interpretación t_24 y m_31 ---")
print(f"t_24 ({t_24}): Ventas totales (en millones) del Producto 2 en la zona Oeste.")
print(f"m_31 ({m_31:.2f}): Ventas esperadas (en millones) del Producto 3 en la zona Norte para el próximo mes.")

# 5. Calcule M.T e interprete m_41^T
M_T = M.T
print("\n--- 5. Matriz M Transpuesta ---")
print(np.round(M_T, 2))
# m_41^T -> Fila 4, Columna 1 de M_T
m_T_41 = M_T[3, 0]
print(f"\nm_41^T ({m_T_41:.1f}): Ventas esperadas (en millones) de la zona Oeste (fila 4 de M_T) para el Producto 1 (columna 1 de M_T).")
```
