# Problema 8

En una sucesión aritmética, la diferencia entre un término y el anterior es $4$ y el décimo término es $-20$. Determine:

1. El primer término de la sucesión.
2. La expresión del término de lugar $n$.
3. Utilice un código en *Python* para calcular el término que ocupa el lugar $100$.
4. Determinar la posición del número $1.680$ en la secuencia. Calcule utilizando un ciclo *while*.

```python
# 1. Encontrar a_1 (matemáticamente)
d = 4
a_10 = -20
# a_n = a_1 + (n - 1) * d  =>  a_1 = a_n - (n - 1) * d
a_1 = a_10 - (10 - 1) * d
print(f"1. El primer término (a_1) es: {a_1}")

# 2. Expresión del término n
# a_n = -56 + (n-1)*4 = -56 + 4n - 4 = 4n - 60
print(f"2. La expresión es: a_n = 4n - 60")

# 3. Calcular a_100
n_100 = 100
a_100 = a_1 + (n_100 - 1) * d
print(f"3. El término que ocupa el lugar 100 es: {a_100}")

# 4. Posición de 1680 (usando ciclo while)
termino_buscado = 1680
n = 0
termino_actual = 0

while termino_actual != termino_buscado:
    n += 1
    termino_actual = a_1 + (n - 1) * d
    if n > 500: # Límite de seguridad
        print("Término no encontrado (límite excedido)")
        break

if termino_actual == termino_buscado:
    print(f"4. El número 1680 ocupa la posición: {n}")
```

# Problema 9

El tercer término de una sucesión geométrica es $5$ y el sexto término es $40$. Determine:

1. La razón geométrica de la sucesión.
2. El primer término de la sucesión.
3. La expresión del término de lugar $n$.
4. Utilice un código en *Python* para calcular el término que ocupa el lugar $8$.
4. ¿Qué lugar ocupa en esta sucesión el término $20.480$? Calcule utilizando un ciclo *while*.

```python
# 1. Encontrar r
a_3 = 5
a_6 = 40
# a_6 = a_3 * r^(6-3) => r = (a_6 / a_3)^(1/3)
r = (a_6 / a_3)**(1/3)
print(f"1. La razón geométrica (r) es: {r}")

# 2. Encontrar a_1
# a_3 = a_1 * r^(3-1) => a_1 = a_3 / r^2
a_1 = a_3 / (r**(3-1))
print(f"2. El primer término (a_1) es: {a_1}")

# 3. Expresión
print(f"3. La expresión es: a_n = {a_1} * {r}**(n-1)")

# 4. Calcular a_8
n_8 = 8
a_8 = a_1 * (r**(n_8 - 1))
print(f"4. El término que ocupa el lugar 8 es: {a_8}")

# 5. Posición de 20480 (usando ciclo while)
termino_buscado = 20480
n = 0
termino_actual = 0

while termino_actual < termino_buscado:
    n += 1
    termino_actual = a_1 * (r**(n - 1))

if abs(termino_actual - termino_buscado) < 0.001: # Tolerancia para decimales
    print(f"5. El número 20480 ocupa la posición: {n}")
else:
    print(f"5. El número 20480 no fue encontrado exactamente.")
```

# Problema 10

Una empresa de tecnología está optimizando sus servidores. Para ello mide las transacciones por segundo (TPS) de sus servidores en cada mes. Se ha obtenido que en el segundo mes, el rendimiento del servidor fue $1,2$ miles de TPS y en el quinto mes, el rendimiento fue $2,7$ miles de TPS. Si el número de TPS que hay mes a mes sigue una sucesión geométrica, determina utilizando un código en *Python*:

1. ¿Cuál es el rendimiento el noveno mes?
2. ¿En qué mes el rendimiento es de $23,4$ miles de TPS?

Utilice dos decimales para el cálculo.

```python
# Datos
a_2 = 1.2
a_5 = 2.7

# Encontrar r y a_1
r = (a_5 / a_2)**(1/3)
a_1 = a_2 / r

# 1. Rendimiento en el noveno mes (a_9)
n_9 = 9
a_9 = a_1 * (r**(n_9 - 1))
# Se redondea el resultado final a 2 decimales, como se pide
print(f"1. El rendimiento el noveno mes es: {a_9:.2f} miles de TPS")

# 2. Mes para 23.4 miles de TPS (usando while)
termino_buscado = 23.4
n = 0
termino_actual = 0

while termino_actual < termino_buscado:
    n += 1
    termino_actual = a_1 * (r**(n - 1))
    
    # Límite de seguridad
    if n > 100:
        print("Término no encontrado (límite excedido)")
        break

# Comprobar si es (aproximadamente) igual
if abs(termino_actual - termino_buscado) < 0.01:
    print(f"2. El rendimiento es de {termino_buscado:.1f} miles de TPS en el mes: {n}")
else:
    # Si no es exacto, mostrar el mes anterior y el actual
    n_anterior = n - 1
    termino_anterior = a_1 * (r**(n_anterior - 1))
    print(f"2. No se alcanza exactamente {termino_buscado:.1f}.")
    print(f"   El mes {n_anterior} da {termino_anterior:.2f} y el mes {n} da {termino_actual:.2f}")
```

# Problema 14

Al llenar un tambor con agua, se registra que durante el minuto $4$ el tambor recibe $128$ litros, y durante el minuto $9$ recibe $88$ litros. Sabiendo que en cada minuto recibe una cantidad $d$ de litros de agua menos que en el minuto anterior, determine:

1. ¿En qué instante el tambor deja de recibir agua?
2. ¿Cuántos litros recibe en total el tambor?

Realice los cálculos anteriores utilizando *Python*.

```python
# Datos
a_4 = 128
a_9 = 88

# Encontrar d y a_1
# a_9 = a_4 + (9-4)d
d = (a_9 - a_4) / 5
# a_4 = a_1 + (4-1)d
a_1 = a_4 - (4 - 1) * d

print(f"Datos calculados: a_1 = {a_1}, d = {d}")

# 1. Instante en que deja de recibir agua (a_n <= 0)
n = 0
termino_actual = a_1

# Buscamos el primer término no positivo
while termino_actual > 0:
    n += 1
    termino_actual = a_1 + (n - 1) * d

n_ultimo_positivo = n - 1
print(f"1. En el minuto {n_ultimo_positivo} recibe {a_1 + (n_ultimo_positivo - 1) * d} litros.")
print(f"   En el minuto {n} recibe {termino_actual} litros.")
print(f"   Por lo tanto, deja de recibir agua en el minuto {n}.")

# 2. Litros totales recibidos (Sumar todos los términos positivos, del 1 al n_ultimo_positivo)
litros_recibidos = []

# Usando el método del formulario
for i in range(n_ultimo_positivo): # Itera de 0 a 19 (para n=1 a n=20)
    n_loop = i + 1
    termino = a_1 + (n_loop - 1) * d
    litros_recibidos.append(termino)

total_litros = sum(litros_recibidos)
print(f"\n2. El tambor recibe en total: {total_litros} litros.")
```
